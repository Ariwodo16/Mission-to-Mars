```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
import time
```

# NASA Mars News

```python
url = 'https://redplanetscience.com/'
browser.visit(url)
time.sleep(1)
html1 = browser.html
soup = bs(html1, "html.parser")
print(soup.prettify)
```

```python
news_title = soup.find('div', class_='content_title').get_text()
news_title
```

```python
news_p = soup.find('div', class_='article_teaser_body').get_text()
news_p
```

## JPL Mars Space Images—Featured Image

```python
url2 = 'https://spaceimages-mars.com'
browser.visit(url2)
full_image = browser.find_by_tag('button')[1]


html = browser.html
soup = bs(html, 'html.parser')

img_start = 'https://spaceimages-mars.com/'
img_end = soup.find_all('a',class_='showimg fancybox-thumbs')[0]['href']
featured_url = img_start+img_end

featured_url
```

# Mars Hemispheres

```python
url4 = 'https://marshemispheres.com/'
browser.visit(url4)

hemi_images = []

links_found = browser.find_by_css('a.product-item img')

for x in range(len(links_found)):
    hemiInfo ={}
    
    browser.find_by_css('a.product-item img')[x].click()
    links = browser.links.find_by_text('Sample').first
    hemiInfo['img_url'] = links['href']
    
    hemiInfo['title'] = browser.find_by_css('h2.title').text
   

    hemi_images.append(hemiInfo)
    
    browser.back()
```
